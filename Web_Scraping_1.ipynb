```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.codewithharry.com/"
```

```python
response=requests.get(url)
response
```

```python
htmlcontent=response.content
#print(htmlcontent)
```

```python
soup=BeautifulSoup(htmlcontent,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
title=soup.title
print(type(title))#Tag
print(type(title.string))#navigableString
print(type(soup))#BeautifulSoup
```

```python
#get the tittle of the html page
title=soup.title
print(title)
```

```python
#get all the paragrapghs from the page
paras=soup.find_all('p')
#print(paras)
```

```python
#get all the anchors tags from the page
anchors=soup.find_all('a')
#print(anchors)
```

```python
#get the first element
print(soup.find('p'))
#get classes of any element in the html page
print(soup.find('p')['class'])
```

```python
#find all the elemnts with class lead
print(soup.find_all("p",class_='lead'))
```

```python
#get the text from the tags/soup
print(soup.find('p').get_text())
```

```python
#get all the links on the page
all_links=set()
for link in anchors:
    if(link.get('href') !='#'):
        linkText="https://www.codewithharry.com/"+link.get('href')
        all_links.add(link)
        print(linkText)
```
